```python
# df = pd.read_csv('/content/cleaned_data.csv')
# print(df)
file_path = '/content/drive/MyDrive/(2022-2024)政大企研所/NCCU DA/112_1_Lndata Project/cleaned_data.csv'
df = pd.read_csv(file_path)
```

```python
df_cleaned_age = df.dropna(subset=['age'])
df_cleaned_age
```

```python
# 定義年齡分類
bins = [18, 30, 40, 50, 60, 99]

# 對年齡分組
df_cleaned_age["age group"] = pd.cut(df_cleaned_age['age'], bins, labels=["18-30歲", "31-40歲", "41-50歲", "51-60歲", "61歲以上"], right=False)

df_cleaned_age
```

```python
# 在各年齡區間下，檢視在性別與品項分類下的「總消費金額」與「平均交易單價」

target_data_1=(
    df_cleaned_age
    .groupby(['age group','categ'], as_index=False)
    .agg({'amount':['sum','mean']})
    # .sort_values(by=('amount','sum'), ascending=False)
)
target_data_1
```

```python
filtered_df = df_cleaned_age[df_cleaned_age['age group'] == '18-30歲']

target_data_1=(
    filtered_df
    .groupby(['categ'], as_index=False)
    .agg({'amount':['sum','mean']})
)
target_data_1
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 假設 target_data_1 已經是您按照年齡組和類別分組，並計算了總和和平均值的 DataFrame

# 準備 x 軸數據，可以是 'age group' 和 'categ' 的組合
# x_data = target_data_1['age group'] + ' - ' + target_data_1['categ']
x_data = target_data_1['categ']


# 準備 'sum' 和 'mean' 的 y 軸數據
y_sum = target_data_1['amount']['sum']
y_mean = target_data_1['amount']['mean']

# 創建帶有雙 y 軸的圖表
fig = make_subplots(specs=[[{"secondary_y": True}]])

# 添加長條圖（用於 'sum'）
fig.add_trace(go.Bar(x=x_data, y=y_sum, name='消費總金額'), secondary_y=False)

# 添加折線圖（用於 'mean'）
fig.add_trace(go.Scatter(x=x_data, y=y_mean, name='平均交易單價', mode='lines+markers'), secondary_y=True)

# 更新 x 軸和主 y 軸的標題
fig.update_xaxes(title_text='各烈酒品項')
fig.update_yaxes(title_text='消費總金額', secondary_y=False)

# 更新次要 y 軸的標題
fig.update_yaxes(title_text='平均交易單價', secondary_y=True)

# 更新布局和標題
fig.update_layout(
    title='18-25歲消費者 消費總金額 與 平均交易單價',
    legend_title='數據類型'
)

# 顯示圖表
fig.show()
```

```python
# 定義年齡組列表
age_groups = ['18-30歲', '31-40歲', '41-50歲', '51-60歲', '61歲以上']

save_path = '/content/drive/MyDrive/(2022-2024)政大企研所/NCCU DA/112_1_Lndata Project/'

# 使用 for 迴圈遍歷每個年齡組
for age_group in age_groups:

  # 根據當前年齡組過濾數據
  filtered_df = df_cleaned_age[df_cleaned_age['age group'] == age_group]

  # 對過濾後的數據進行分組和聚合
  target_data = (
      filtered_df
      .groupby(['categ'], as_index=False)
      .agg({'amount': ['sum', 'mean']})
  )

  # 準備繪圖數據
  x_data = target_data['categ']
  y_sum = target_data['amount']['sum']
  y_mean = target_data['amount']['mean']

  # 創建帶有雙 y 軸的圖表
  fig = make_subplots(specs=[[{"secondary_y": True}]])

  # 添加長條圖和折線圖
  fig.add_trace(go.Bar(x=x_data, y=y_sum, name='消費總金額'), secondary_y=False)
  fig.add_trace(go.Scatter(x=x_data, y=y_mean, name='平均交易單價', mode='lines+markers'), secondary_y=True)

  # 更新圖表標題和軸標籤
  fig.update_layout(
      font=dict(family="Noto Sans CJK JP"),
      title=f'{age_group}消費者 消費總金額 與 平均交易單價',
      legend_title='數據類型',
      xaxis_title='各烈酒品項',
      yaxis_title='消費總金額',
      yaxis2_title='平均交易單價'
  )

  # 保存圖表為 PNG 文件
  file_name = f'{save_path}plot_{age_group}.png'
  fig.write_image(file_name)

  # 顯示圖表
  fig.show()
```
